```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd

# Load the dataset
data_path = 'Global YouTube Statistics.csv'
data = pd.read_csv(data_path, encoding='latin1')

# Set plot style
sns.set_theme(style="whitegrid")

# 1. Bar Chart: Top 10 YouTubers by Subscribers
plt.figure(figsize=(10, 6))
top_youtubers = data.nlargest(10, 'subscribers')
sns.barplot(x='subscribers', y='Youtuber', data=top_youtubers, palette='viridis')
plt.title('Top 10 YouTubers by Subscribers')
plt.xlabel('Subscribers (in billions)')
plt.ylabel('Youtuber')
plt.tight_layout()
plt.show()

# 2. Histogram: Distribution of Video Views
plt.figure(figsize=(10, 6))
sns.histplot(data['video views'], bins=30, kde=True, color='blue')
plt.title('Distribution of Video Views')
plt.xlabel('Video Views (in billions)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# 3. Scatter Plot: Subscribers vs. Video Views
plt.figure(figsize=(10, 6))
sns.scatterplot(x='subscribers', y='video views', hue='category', data=data, palette='deep', alpha=0.7)
plt.title('Subscribers vs. Video Views by Category')
plt.xlabel('Subscribers (in billions)')
plt.ylabel('Video Views (in billions)')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

'''# 4. Heatmap: Correlation Matrix
plt.figure(figsize=(12, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', cbar=True)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()'''

```

```python
# 6. Boxplot: Monthly Earnings by Channel Type
plt.figure(figsize=(10, 6))
sns.boxplot(x='channel_type', y='highest_monthly_earnings', data=data, palette='coolwarm')
plt.title('Monthly Earnings by Channel Type')
plt.xlabel('Channel Type')
plt.ylabel('Highest Monthly Earnings (in USD)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 7. Line Plot: Average Subscribers Growth Over Time
plt.figure(figsize=(10, 6))
data['created_year'] = data['created_year'].fillna(0).astype(int)
avg_subscribers_by_year = data.groupby('created_year')['subscribers'].mean()
avg_subscribers_by_year.plot(kind='line', marker='o', color='purple')
plt.title('Average Subscribers Growth Over Time')
plt.xlabel('Year')
plt.ylabel('Average Subscribers (in billions)')
plt.tight_layout()
plt.show()

# 8. Count Plot: Most Common Categories
plt.figure(figsize=(10, 6))
sns.countplot(y='category', data=data, order=data['category'].value_counts().index, palette='muted')
plt.title('Most Common Categories')
plt.xlabel('Count')
plt.ylabel('Category')
plt.tight_layout()
plt.show()

# 9. Violin Plot: Earnings Distribution by Region
plt.figure(figsize=(10, 6))
data['Region'] = data['Country'].apply(lambda x: 'Unknown' if pd.isnull(x) else x)
sns.violinplot(x='Region', y='highest_yearly_earnings', data=data, palette='pastel', cut=0)
plt.title('Yearly Earnings Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Highest Yearly Earnings (in USD)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 10. Pairplot: Relationships Between Key Metrics
plt.figure(figsize=(10, 6))
sns.pairplot(data, vars=['subscribers', 'video views', 'uploads', 'highest_monthly_earnings'], hue='category', palette='husl', diag_kind='kde')
plt.suptitle('Relationships Between Key Metrics', y=1.02)
plt.tight_layout()
plt.show()
```
